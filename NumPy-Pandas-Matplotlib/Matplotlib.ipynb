# 1 PyPlot
PyPlot is a submodule of Matplotlib that provides a MATLAB-like interface for plotting. Usually we import it as plt.

## Drawing x and y points
The plot() function draw a line from point to point.  
Argument 1: An array containing the x coordinates of the points.  
Argument 2: An array containing the y coordinates of the points.

```python
import matplotlib.pyplot as plt
```

```python
# draw a straight line from (0, 0) to (6, 250)
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints,ypoints)
plt.show()
```

## Drawing without lines
To draw only the markers, we can use the keyword string argument 'o', which means circles.

```python
plt.plot(xpoints,ypoints, 'o')
plt.show()
```

## Multiple points
We can plot any number of points in a diagram, just make sure the number of points in both arrays is the same.

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints,ypoints)
plt.show()
```

## Default x points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc., depending on the length of the y points).

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

# 2 Markers
## Keyword argument marker
You can use the keyword argument marker to emphasize each point with a specified marker.

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o')
plt.show()
```

```python
plt.plot(ypoints, marker='*')
plt.show()
```

## Marker Size Color Reference
Make good use of search

## Formatted String fmt
Shortcut string notation parameter fmt to specify the marker: marker|line|color

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, 'o:r')
plt.show()
```

## Marker size

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o', ms=20)
plt.show()
```

## Marker color

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o', ms=20, mec='r') # markeredgecolor
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o', ms=20, mfc='y') # markerfacecolor
plt.show()
```

```python
# We can also use hex color values and color names.
plt.plot(ypoints, marker='o', ms=20, mec='#800080', mfc='hotpink')
```

# 3 Lines
## Line Style

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle='dotted')
plt.show()
```

```python
plt.plot(ypoints, linestyle='dashed')
plt.show()
```

## Shorter Syntax
linestyle --> ls  
dotted --> :  
dashed --> --
