# 1 PyPlot
PyPlot is a submodule of Matplotlib that provides a MATLAB-like interface for plotting. Usually we import it as plt.

## Drawing x and y points
The plot() function draw a line from point to point.  
Argument 1: An array containing the x coordinates of the points.  
Argument 2: An array containing the y coordinates of the points.

```python
import matplotlib.pyplot as plt
```

```python
# draw a straight line from (0, 0) to (6, 250)
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints,ypoints)
plt.show()
```

## Drawing without lines
To draw only the markers, we can use the keyword string argument 'o', which means circles.

```python
plt.plot(xpoints,ypoints, 'o')
plt.show()
```

## Multiple points
We can plot any number of points in a diagram, just make sure the number of points in both arrays is the same.

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints,ypoints)
plt.show()
```

## Default x points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc., depending on the length of the y points).

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

# 2 Markers
## Keyword argument marker
You can use the keyword argument marker to emphasize each point with a specified marker.

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o')
plt.show()
```

```python
plt.plot(ypoints, marker='*')
plt.show()
```

Make good use of search

## Formatted String fmt
Shortcut string notation parameter fmt to specify the marker: marker|line|color

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, 'o:r')
plt.show()
```

## Marker size

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o', ms=20)
plt.show()
```

## Marker color

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o', ms=20, mec='r') # markeredgecolor
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o', ms=20, mfc='y') # markerfacecolor
plt.show()
```

```python
# We can also use hex color values and color names.
plt.plot(ypoints, marker='o', ms=20, mec='#800080', mfc='hotpink')
```

## Multiple Lines
Adding more lines by simply calling the plt.plot() function multiple times to plot multiple lines.

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.show()
```

# 4 Labels and Titles
## Create Labels for the Plots
Using PyPlot, you can use the xlabel() and ylabel() functions to set labels for the x-axis and y-axis.

```python
import matplotlib

# Set the default font for sans-serif text to 'DejaVu Sans'
matplotlib.rcParams['font.sans-serif'] = ['DejaVu Sans']

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

## Create Titles for the Plots
Using Pyplot, you can use the title() function to set a title for the plot.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

## Set Font Properties for Titles and Labels
Use the xlabel(), ylabel(), and title() functions with the fontdict parameter to set the font properties for titles and labels.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family': 'serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'serif', 'color': 'darkred', 'size': 15}

plt.title("Sports Watch Data", fontdict=font1)
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)

plt.plot(x, y)
plt.show()
```

## Title Positions
Use the loc parameter in the title() function to position the title.
Valid values are: 'left', 'right', and 'center'. The default is 'center'.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc='left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

# 5 Grid

Using Pyplot, use the grid() function to add grid lines to the plot.

# 6 Multiple Plots
## Display Multiple Plots
Using the subplots() function, you can draw multiple plots in one figure:

```python
# Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) # 1 row, 2 columns, first subplot
plt.plot(x, y)

# Plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) # 1 row, 2 columns, second subplot
plt.plot(x, y)

plt.show()
```

## subplots()
rows, columns, and the index

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x, y)

plt.subplot(2, 3, 2)
plt.plot(x, y)

plt.subplot(2, 3, 3)
plt.plot(x, y)

plt.subplot(2, 3, 4)
plt.plot(x, y)

plt.subplot(2, 3, 5)
plt.plot(x, y)

plt.subplot(2, 3, 6)
plt.plot(x, y)

plt.show()
```

## Title
Using the title() function to add titles to each plot:

```python
# plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('SALES')

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('INCOME')

plt.show()
```

## Super Title
You can use the suptitle() function to add a title to the entire figure.

```python
# plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('SALES')

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('INCOME')

plt.suptitle('MY SHOP')
plt.show()
```

# 7 Scatter Plot
## Creating a Scatter Plot
Using Pyplot, you can use the scatter() function to create scatter plots.
The scatter() function creates a scatter plot with several points. It requires two arrays of the same length, one for the x-axis values and one for the y-axis values:

```python
x = np.array([5, 7, 8, 7, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77])

plt.scatter(x, y)
plt.show()
```

## Color
Using the color parameter to set the color of the points in each scatter plot:

```python
x = np.array([5, 7, 8, 7, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77])
plt.scatter(x, y, color='hotpink')

x = np.array([2, 2, 8, 1, 5, 8])
y = np.array([100, 105, 84, 105, 90, 99])
plt.scatter(x, y, color='#88c999')

plt.show()
```

## Coloring Each Point
You can specify a list of colors for each point using the c parameter instead of the color parameter:

```python
x = np.array([5, 7, 8, 7, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77])
colors = np.array(['red', 'green', 'blue', 'yellow', 'pink', 'black', 'orange', 'purple', 'beige', 'brown', 'gray'])

plt.scatter(x, y, c=colors)
plt.show()
```

## Color Maps
The Matplotlib module has many available color maps.
A color map looks like a range of colors, where each color corresponds to a value in the range from 0 to 100.

### How to Use Color Maps
You can use the cmap keyword argument to specify a color map. In this example, it is set to 'viridis', which is one of the built-in color maps in Matplotlib.
Additionally, you need to create an array of values (ranging from 0 to 100). Each point in the scatter plot will have a value from this array:

```python
# Use the plt.colorbar() statement to include a color bar in the plot
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()

plt.show()
```

##  Available Color Maps
You can select built-in color maps in Matplotlib.

##  Size
Use the s parameter to change the size of the points.
Similar to colors, make sure the length of the sizes array is the same as the lengths of the x and y arrays:

```python
x = np.array([1, 2, 5, 7, 8, 7, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
sizes = np.array([20, 50, 100, 200, 500, 1000, 60, 90, 10, 300, 600, 800, 75])

plt.scatter(x, y, s=sizes)

plt.show()
```

## Transparency
You can use the alpha parameter to adjust the transparency of the points.
Similar to colors, make sure the length of the sizes array is the same as the lengths of the x and y arrays:

```python
x = np.array([1, 2, 5, 7, 8, 7, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
sizes = np.array([20, 50, 100, 200, 500, 1000, 60, 90, 10, 300, 600, 800, 75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

## Combining Color, Size, and Transparency
A scatter plot with points of different colors, sizes, and transparency:

```python
x = np.random.randint(100, size=100)
y = np.random.randint(100, size=100)
colors = np.random.randint(100, size=100)
sizes = 10 * np.random.randint(100, size=100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()

plt.show()
```

# 8 Bar Charts

In the Pyplot package, use the bar() function to draw bar charts:

# 10 Pie Chart
## Creating a Pie Chart
In the Pyplot package, you can use the pie() function to create a pie chart:

```python
y = np.array([35, 25, 25, 15]) # the sum should be 100

plt.pie(y)
plt.show()
```

## Labels
Use the label parameter to add labels to the pie chart.
The label parameter should be a list, with each wedge having a label:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels)
plt.show()
```

## Starting Angle
The default starting angle is aligned with the x-axis, but you can change the starting angle using the startangle parameter.
The startangle parameter is defined in degrees, with a default angle of 0:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels, startangle=90)
plt.show()
```

## Explode
Would you like to pull a slice out of the pie? The explode parameter allows you to do this.  
If you specify the explode parameter and it is not None, it must be an array where each wedge has a value.  
Each value represents how far the wedge is from the center of the pie chart:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels=mylabels, explode=myexplode)
plt.show()
```

## Shadow
By setting the shadow parameter to True, you can add a shadow to the pie chart:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels=mylabels, explode=myexplode, shadow=True)
plt.show()
```

## Colors
Use the colors parameter to set the colors of the wedges.  
If you specify the colors parameter, it must be an array where each wedge has a value:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels=mylabels, colors=mycolors)
plt.show()
```

## Legend
To add a legend to each wedge of the pie chart, use the legend() function:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels)
plt.legend()
plt.show()
```

### Legend with Title
To add a title to the legend, pass the title parameter to the legend() function:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels)
plt.legend(title="Four Fruits")
plt.show()
```
